```python
import numpy as np
import pandas as pd
```

```python
# Define the dataset, only use relevant columns.
df = pd.read_csv('Guardian_reviews_ready_relevant_info.csv', usecols = ['review_text', 'stars'])

```

```python
# Oversampling (1 stars translated to Japanese and back to English) Setting up the csv properly again 

df_trans1stars = pd.read_csv('df_1_stars(backtranslated).csv', sep='\t', lineterminator='\r', header=None)

# Strip from n/
df_trans1stars[0] = df_trans1stars[0].str.strip()

# Make a new column
df_trans1stars['review_text'] = df_trans1stars[0]

#Drop old column
df_trans1stars = df_trans1stars.drop([0], axis=1)

# Create column stars with int 1
df_trans1stars['stars'] = 1

df = pd.concat([df_trans1stars, df])
#df = df.drop([0], axis=1)

df['stars'].hist();df.head()
df.info()

df_count = df.groupby('stars')['review_text'].count()
print(df_count)

df.head(20)
```

```python
# Oversampling (2 stars translated to Japanese and back to English) Setting up the csv properly again 
df_trans2stars = pd.read_csv('df_2_stars(backtranslated).csv', sep='\t', lineterminator='\r', header=None)

# Strip from n/
df_trans2stars[0] = df_trans2stars[0].str.strip()

# Make a new column
df_trans2stars['review_text'] = df_trans2stars[0]

#Drop old column
df_trans2stars = df_trans2stars.drop([0], axis=1)

# Create column stars with int 2
df_trans2stars['stars'] = 2

df = pd.concat([df_trans2stars, df])

df['stars'].hist();df.head()
df.info()

df_count = df.groupby('stars')['review_text'].count()
print(df_count)

df.head(20)
```

```python
# Oversampling (5 stars translated to Japanese and back to English) Setting up the csv properly again 
df_trans5stars = pd.read_csv('df_5_stars(backtranslated).csv', sep='\t', lineterminator='\r', header=None)


# Strip from n/
df_trans5stars[0] = df_trans5stars[0].str.strip()

# Make a new column
df_trans5stars['review_text'] = df_trans5stars[0]

#Drop old column
df_trans5stars = df_trans5stars.drop([0], axis=1)

# Create column stars with int 5
df_trans5stars['stars'] = 5

df = pd.concat([df_trans5stars, df])

df['stars'].hist();df.head()
df.info()

df_count = df.groupby('stars')['review_text'].count()
print(df_count)

df.head(20)
#df_trans5stars.head(20)
```

```python
# Oversampling - Adding 2 and 5 stars from NME reviews
df_NME = pd.read_csv('NME_reviews_ready.csv', usecols = ['review_text', 'stars'])

# Drop duplicates
df_NME = df_NME.drop_duplicates()

# to change use .astype() 
df_NME['stars'] = df_NME.stars.astype(int)

# Get only 5 star reviews
df_NME_5stars = df_NME.loc[df_NME['stars'] == 10]
#Change the value 10 out of 10 to 5 out of 5
df_NME_5stars = df_NME_5stars.replace(10, 5)

# Get only 2 star reviews
df_NME_2stars = df_NME.loc[df_NME['stars'] == 4]
#Change the value 10 out of 10 to 5 out of 5
df_NME_2stars = df_NME_2stars.replace(4, 2)
print(df_NME_2stars)

#df_count = df_NME.groupby('stars')['review_text'].count()
#print(df_count)

# Concatenate dataframes with original dataset
df = pd.concat([df_NME_5stars, df_NME_2stars, df])

df_count = df.groupby('stars')['review_text'].count()
print(df_count)

df['stars'].hist();df.head()
df.info()
```

```python
df_unique = df.groupby('stars')['review_text'].nunique()
print(df_unique)
```

```python
df = df.drop_duplicates(subset='review_text', keep='first')
df_count = df.groupby('stars')['review_text'].count()
print(df_count)
```

```python
#df.to_csv(r'Guardian_reviews_ready_oversampled.csv')
```

```python
# Undersampling after Oversampling

# Shuffle the Dataset.
shuffled_df = df.sample(frac=1,random_state=4)

#Randomly select maximum X number of samples
df_1_stars = shuffled_df.loc[shuffled_df['stars'] == 1].sample(n=188,random_state=42)

df_2_stars = shuffled_df.loc[shuffled_df['stars'] == 2].sample(n=1640,random_state=42)

df_3_stars = shuffled_df.loc[shuffled_df['stars'] == 3].sample(n=2549,random_state=42)

df_4_stars = shuffled_df.loc[shuffled_df['stars'] == 4].sample(n=2549,random_state=42)

df_5_stars = shuffled_df.loc[shuffled_df['stars'] == 5].sample(n=2549,random_state=42)

# Concatenate both dataframes again
df = pd.concat([df_1_stars, df_2_stars, df_3_stars, df_4_stars, df_5_stars])

df['stars'].hist();df.head()
```
